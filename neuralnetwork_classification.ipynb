```python
import sklearn
```

```python
from sklearn.datasets import make_circles

#sample data
n_samples=1000

#create circles
X,y=make_circles(n_samples,
                 noise=0.03,
                 random_state=42)
```

```python
len(X),len(y)
```

```python
print(f"First 5 samples of X :{X[:5]}")
print(f"first 5 smaples of y :{y[:5]}")
```

```python
#make dataframe of circle data
import pandas as pd
circles=pd.DataFrame({"X1":X[:,0],
                      "X2":X[:,1],
                      "label":y})
circles.head(10)
```
